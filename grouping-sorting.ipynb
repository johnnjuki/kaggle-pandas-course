```python
import pandas as pd
```

```python
reviews = pd.read_csv("./data/winemag-data-130k-v2.csv", index_col=0)
pd.set_option('display.max_rows', 5)
reviews
```

## Groupwise analysis



```python
# We can replicate what `value_counts()` does by doing:
reviews.groupby('points').points.count()
```

```python
reviews.groupby('points').groups
```

```python
reviews.groupby('points').get_group(87)
```

```python
reviews.groupby('points').describe()
```

```python
reviews.groupby('points').size()
```

```python
# Get the cheapest wine in each point value category
reviews.groupby('points').price.min()
```

```python
reviews
```

```python
# Select the name of the first wine reviewed from each winery
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# Pick out the best wine by country and province
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
reviews.points.idxmax()
```

```python
reviews.iloc[345].points
```

`agg()`, a `groupby()` method, lets you do a bunch of different functions on your DataFrame simultaneously.

```python
# Generate a simple statistical summary
reviews.groupby(['country']).price.agg([len, min, max])
```

## Mulit-indexes

A multi-index differs from a regular index in that it hsa multiple levels. For example:


```python
countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
mi = countries_reviewed.index
type(mi)
```

```python
# Convert back to a regular index
countries_reviewed.reset_index()
```

## Sorting

Looking again at `countries_reviewed` we can see that grouping returns data in index order, not in value order. That is to say, when outputting the result of a `groupby`, the order of the rows is dependent on the values in the index, not in the data.

To get data in the order want it in we can sort it ourselves. The `sort_values()` method is handy for this.

```python
countries_reviewed = countries_reviewed.reset_index()
countries_reviewed.sort_values(by='len')
```

```python
# Sort in descending order
countries_reviewed.sort_values(by='len', ascending=False)
```

To sort by index values, use the companion method `sort_index()`. 

```python
countries_reviewed.sort_index()
```

```python
# Sort by more than one column at a time
countries_reviewed.sort_values(by=['country', 'len'])
```
